```python
# dfca: DataFrame Combiner Algorithm
def dfca(thedata1, thedata2, thedata3):
    thedata1 = thedata1
    thedata2 = thedata2
    thedata3 = thedata3
    
    thedata4 = pd.merge(thedata1, thedata2, how='left', left_on='customer_id', right_on='clientid')
    
    
    df4 = thedata4
    
    #ttc: 
    ttc = thedata4.shape[0]
    
    df4['ttc'] = ttc 
    
    # tepp: The email prov
    df4['tepp'] = df4['email'].str.split('@').str[1]
    df4['tepp'] = df4['tepp'].str.split('.').str[0]
    df4['tepp'] = df4['tepp'].str.lower()
    
    df4['tepp'] = df4['tepp'].fillna('noemail')
    
    df4['geo_x'] = df4['geo_x'].fillna('NC')
    df4['geo_y'] = df4['geo_y'].fillna('NC')
    

    
    # bgc: Binerized Geographical Consistency
    df4['bgc'] = (df4['geo_x'] == df4['geo_y']).astype(int)
    
   
    
    # ikwn: is Keyword Null 
    df4['ikwn'] = (df4['kw'].isnull() == True).astype(int)
    
    df4['returns'] = df4['returns'].fillna(0)
    
    df4['returns'] = df4['returns'].astype(float)
    
    df4['returns'] = pd.to_numeric(df4['returns'], errors='raise')
    
    
    # tba: Target Binarized Algorithm
    def tba(state):
        if state == 'ChargeBack':
            return 1
        elif state == 'Refund':
            return 1
        elif state == 'Reversed':
            return 1
        else:
            return 0 
    
   
   
    
  
   
    
    # icb: is Charge Back 
    df4['icb'] = df4['state'].apply(tba)
    
    
    
    # tcbgcn: Total Customers' Binarized Geographical Consistency Number 
    df4['tcbgcn'] =  df4.groupby('customer_id')['icb'].sum()
    
    # tccbp: The Customer Charge Back Probability 
    df4['tccbp'] = df4['tcbgcn'] / df4[df4['icb'] == 1].shape[0]
    df4['tccbp'] = df4['tccbp'].fillna(0)
    
    
    
    df4cb = df4[df4['icb'] == 1]
    
    
    totaltepp = df4.groupby('tepp')['icb'].count().to_frame().reset_index()
    totaltepp = totaltepp.rename(columns={'icb':'total_tepp'})
    
    
    cbtepp = df4.groupby('tepp')['icb'].sum().to_frame().reset_index()
    cbtepp = cbtepp.rename(columns={'icb':'cb_tepp'})
    
    
    combinedtepp = pd.merge(cbtepp, totaltepp, how='inner', left_on = 'tepp', right_on = 'tepp')
    combinedtepp['cb_tepp_pc'] =  combinedtepp['cb_tepp'] / combinedtepp['total_tepp']
    
    combinedtepp = combinedtepp[['tepp', 'cb_tepp_pc']]
    
    df4 = df4.merge(combinedtepp, how='inner', left_on = 'tepp', right_on = 'tepp')
    
    
    totalitem = df4.groupby('item_name')['icb'].count().to_frame().reset_index()
    totalitem = totalitem.rename(columns={'icb':'total_item'})
    
    
    cbitem = df4.groupby('item_name')['icb'].sum().to_frame().reset_index()
    cbitem = cbitem.rename(columns={'icb':'cb_item'})
    
    
    combineditem = pd.merge(cbitem, totalitem, how='inner', left_on = 'item_name', right_on = 'item_name')
    combineditem['cb_item_pc'] =  combineditem['cb_item'] / combineditem['total_item']
    
    combineditem = combineditem[['item_name', 'cb_item_pc']]
    
    df4 = df4.merge(combineditem, how='inner', left_on = 'item_name', right_on = 'item_name')
    
    
    totalprovider = df4.groupby('provider')['icb'].count().to_frame().reset_index()
    totalprovider = totalprovider.rename(columns={'icb':'total_provider'})
    
    
    cbprovider = df4.groupby('provider')['icb'].sum().to_frame().reset_index()
    cbprovider = cbprovider.rename(columns={'icb':'cb_provider'})
    
    
    combinedprovider = pd.merge(cbprovider, totalprovider, how='inner', left_on = 'provider', right_on = 'provider')
    combinedprovider['cb_provider_pc'] =  combinedprovider['cb_provider'] / combinedprovider['total_provider']
    
    combinedprovider = combinedprovider[['provider', 'cb_provider_pc']]
    
    df4 = df4.merge(combinedprovider, how='inner', left_on = 'provider', right_on = 'provider')
    
    
    totalcampaign = df4.groupby('campaign')['icb'].count().to_frame().reset_index()
    totalcampaign = totalcampaign.rename(columns={'icb':'total_campaign'})
    
    
    cbcampaign = df4.groupby('campaign')['icb'].sum().to_frame().reset_index()
    cbcampaign = cbcampaign.rename(columns={'icb':'cb_campaign'})
    
    
    combinedcampaign = pd.merge(cbcampaign, totalcampaign, how='inner', left_on = 'campaign', right_on = 'campaign')
    combinedcampaign['cb_campaign_pc'] =  combinedcampaign['cb_campaign'] / combinedcampaign['total_campaign']
    
    combinedcampaign = combinedcampaign[['campaign', 'cb_campaign_pc']]
    
    df4 = df4.merge(combinedcampaign, how='inner', left_on = 'campaign', right_on = 'campaign')
    
    df4['descriptor'] = df4['descriptor'].fillna('nodescriptor')
    df4['descriptor'] = df4['descriptor'].str.lower()
    df4['descriptor'] = df4['descriptor'].str.replace('pcappstore:fast', 'fast')

    
    totaldescriptor = df4.groupby('descriptor')['icb'].count().to_frame().reset_index()
    totaldescriptor = totaldescriptor.rename(columns={'icb':'total_descriptor'})
    
    
    cbdescriptor = df4.groupby('descriptor')['icb'].sum().to_frame().reset_index()
    cbdescriptor = cbdescriptor.rename(columns={'icb':'cb_descriptor'})
    
    
    combineddescriptor = pd.merge(cbdescriptor, totaldescriptor, how='inner', left_on = 'descriptor', right_on = 'descriptor')
    combineddescriptor['cb_descriptor_pc'] =  combineddescriptor['cb_descriptor'] / combineddescriptor['total_descriptor']
    
    combineddescriptor = combineddescriptor[['descriptor', 'cb_descriptor_pc']]
    
    df4 = df4.merge(combineddescriptor, how='inner', left_on = 'descriptor', right_on = 'descriptor')
    
    df4['referrer'] = df4['referrer'].fillna('noreferrer')
    
    df4['referrer'] = df4['referrer'].str.split('.')
    df4['referrer'] = df4['referrer'].str[1]
    
    df4['referrer'] = df4['referrer'].fillna('noreferrer')
    
    
    totalreferrer = df4.groupby('referrer')['icb'].count().to_frame().reset_index()
    totalreferrer = totalreferrer.rename(columns={'icb':'total_referrer'})
    
    
    cbreferrer = df4.groupby('referrer')['icb'].sum().to_frame().reset_index()
    cbreferrer = cbreferrer.rename(columns={'icb':'cb_referrer'})
    
    
    combinedreferrer = pd.merge(cbreferrer, totalreferrer, how='inner', left_on = 'referrer', right_on = 'referrer')
    combinedreferrer['cb_referrer_pc'] =  combinedreferrer['cb_referrer'] / combinedreferrer['total_referrer']
    
    combinedreferrer = combinedreferrer[['referrer', 'cb_referrer_pc']]
    
    df4 = df4.merge(combinedreferrer, how='inner', left_on = 'referrer', right_on = 'referrer')
    
    
    df4['src_domain'] = df4['src_domain'].fillna('nodomain')
    
    totalsrc_domain = df4.groupby('src_domain')['icb'].count().to_frame().reset_index()
    totalsrc_domain = totalsrc_domain.rename(columns={'icb':'total_src_domain'})
    
    
    cbsrc_domain = df4.groupby('src_domain')['icb'].sum().to_frame().reset_index()
    cbsrc_domain = cbsrc_domain.rename(columns={'icb':'cb_src_domain'})
    
    
    combinedsrc_domain = pd.merge(cbsrc_domain, totalsrc_domain, how='inner', left_on = 'src_domain', right_on = 'src_domain')
    combinedsrc_domain['cb_src_domain_pc'] =  combinedsrc_domain['cb_src_domain'] / combinedsrc_domain['total_src_domain']
    
    combinedsrc_domain = combinedsrc_domain[['src_domain', 'cb_src_domain_pc']]
    
    df4 = df4.merge(combinedsrc_domain, how='inner', left_on = 'src_domain', right_on = 'src_domain')
    
    df4['ai'] = df4['ai'].fillna('noai')
    
    totalai = df4.groupby('ai')['icb'].count().to_frame().reset_index()
    totalai = totalai.rename(columns={'icb':'total_ai'})
    
    
    cbai = df4.groupby('ai')['icb'].sum().to_frame().reset_index()
    cbai = cbai.rename(columns={'icb':'cb_ai'})
    
    
    combinedai = pd.merge(cbai, totalai, how='inner', left_on = 'ai', right_on = 'ai')
    combinedai['cb_ai_pc'] =  combinedai['cb_ai'] / combinedai['total_ai']
    
    combinedai = combinedai[['ai', 'cb_ai_pc']]
    
    df4 = df4.merge(combinedai, how='inner', left_on = 'ai', right_on = 'ai')
    
    df4['ip_state_name'] = df4['ip_state_name'].fillna('no_ip_state_name')
    
    totalip_state_name = df4.groupby('ip_state_name')['icb'].count().to_frame().reset_index()
    totalip_state_name = totalip_state_name.rename(columns={'icb':'total_ip_state_name'})
    
    
    cbip_state_name = df4.groupby('ip_state_name')['icb'].sum().to_frame().reset_index()
    cbip_state_name = cbip_state_name.rename(columns={'icb':'cb_ip_state_name'})
    
    
    combinedip_state_name = pd.merge(cbip_state_name, totalip_state_name, how='inner', left_on = 'ip_state_name', right_on = 'ip_state_name')
    combinedip_state_name['cb_ip_state_name_pc'] =  combinedip_state_name['cb_ip_state_name'] / combinedip_state_name['total_ip_state_name']
    
    combinedip_state_name = combinedip_state_name[['ip_state_name', 'cb_ip_state_name_pc']]
    
    df4 = df4.merge(combinedip_state_name, how='inner', left_on = 'ip_state_name', right_on = 'ip_state_name')
    
    df4['ip_city'] = df4['ip_city'].fillna('noipcity')
    
    totalip_city = df4.groupby('ip_city')['icb'].count().to_frame().reset_index()
    totalip_city = totalip_city.rename(columns={'icb':'total_ip_city'})
    
    
    cbip_city = df4.groupby('ip_city')['icb'].sum().to_frame().reset_index()
    cbip_city = cbip_city.rename(columns={'icb':'cb_ip_city'})
    
    
    combinedip_city = pd.merge(cbip_city, totalip_city, how='inner', left_on = 'ip_city', right_on = 'ip_city')
    combinedip_city['cb_ip_city_pc'] =  combinedip_city['cb_ip_city'] / combinedip_city['total_ip_city']
    
    combinedip_city = combinedip_city[['ip_city', 'cb_ip_city_pc']]
    
    df4 = df4.merge(combinedip_city, how='inner', left_on = 'ip_city', right_on = 'ip_city')
    
    
    df4['item_key'] = df4['item_key'].fillna('noipkey')
    
    totalitem_key = df4.groupby('item_key')['icb'].count().to_frame().reset_index()
    totalitem_key = totalitem_key.rename(columns={'icb':'total_item_key'})
    
    
    cbitem_key = df4.groupby('item_key')['icb'].sum().to_frame().reset_index()
    cbitem_key = cbitem_key.rename(columns={'icb':'cb_item_key'})
    
    
    combineditem_key = pd.merge(cbitem_key, totalitem_key, how='inner', left_on = 'item_key', right_on = 'item_key')
    combineditem_key['cb_item_key_pc'] =  combineditem_key['cb_item_key'] / combineditem_key['total_item_key']
    
    combineditem_key = combineditem_key[['item_key', 'cb_item_key_pc']]
    
    df4 = df4.merge(combineditem_key, how='inner', left_on = 'item_key', right_on = 'item_key')
    
    
    df4['strategy_id'] = df4['strategy_id'].fillna('nostrategyid')
    
    totalstrategy_id = df4.groupby('strategy_id')['icb'].count().to_frame().reset_index()
    totalstrategy_id = totalstrategy_id.rename(columns={'icb':'total_strategy_id'})
    
    
    cbstrategy_id = df4.groupby('strategy_id')['icb'].sum().to_frame().reset_index()
    cbstrategy_id = cbstrategy_id.rename(columns={'icb':'cb_strategy_id'})
    
    
    combinedstrategy_id = pd.merge(cbstrategy_id, totalstrategy_id, how='inner', left_on = 'strategy_id', right_on = 'strategy_id')
    combinedstrategy_id['cb_strategy_id_pc'] =  combinedstrategy_id['cb_strategy_id'] / combinedstrategy_id['total_strategy_id']
    
    combinedstrategy_id = combinedstrategy_id[['strategy_id', 'cb_strategy_id_pc']]
    
    df4 = df4.merge(combinedstrategy_id, how='inner', left_on = 'strategy_id', right_on = 'strategy_id')
    
    
    df4['at'] = df4['at'].fillna('noat')
    
    totalat = df4.groupby('at')['icb'].count().to_frame().reset_index()
    totalat = totalat.rename(columns={'icb':'total_at'})
    
    
    cbat = df4.groupby('at')['icb'].sum().to_frame().reset_index()
    cbat = cbat.rename(columns={'icb':'cb_at'})
    
    
    combinedat = pd.merge(cbat, totalat, how='inner', left_on = 'at', right_on = 'at')
    combinedat['cb_at_pc'] =  combinedat['cb_at'] / combinedat['total_at']
    
    combinedat = combinedat[['at', 'cb_at_pc']]
    
    df4 = df4.merge(combinedat, how='inner', left_on = 'at', right_on = 'at')
    
    df4['campaign_first'] = df4['campaign_first'].fillna('nocampaign')
    df4['campaign_last'] = df4['campaign_last'].fillna('nocampaign')
    
    # afalctc: Are First and Lst Campaigns are the Same 
   
    # tcba: The Campaign Comprison and Matching Binarization Algorithm 
    def tccamba(campaign_first, campaign_last):
        
        if campaign_first == campaign_last:
            return 1
        else:
            return 0 
        
        
    # afalcts: Are First and Last Campight the Same
    df4['afalcts'] = df4.apply(lambda x: tccamba(x['campaign_first'], x['campaign_last']), axis=1)
    
    
    df4['campaign_first'] = df4['campaign_first'].fillna('nocampaignfirst')
    
    totalcampaign_first = df4.groupby('campaign_first')['icb'].count().to_frame().reset_index()
    totalcampaign_first = totalcampaign_first.rename(columns={'icb':'total_campaign_first'})
    
    
    cbcampaign_first = df4.groupby('campaign_first')['icb'].sum().to_frame().reset_index()
    cbcampaign_first = cbcampaign_first.rename(columns={'icb':'cb_campaign_first'})
    
    
    combinedcampaign_first = pd.merge(cbcampaign_first, totalcampaign_first, how='inner', left_on = 'campaign_first', right_on = 'campaign_first')
    combinedcampaign_first['cb_campaign_first_pc'] =  combinedcampaign_first['cb_campaign_first'] / combinedcampaign_first['total_campaign_first']
    
    combinedcampaign_first = combinedcampaign_first[['campaign_first', 'cb_campaign_first_pc']]
    
    df4 = df4.merge(combinedcampaign_first, how='inner', left_on = 'campaign_first', right_on = 'campaign_first')
    
    
    
    df4['campaign_last'] = df4['campaign_last'].fillna('nocampaignlast')
    
    totalcampaign_last = df4.groupby('campaign_last')['icb'].count().to_frame().reset_index()
    totalcampaign_last = totalcampaign_last.rename(columns={'icb':'total_campaign_last'})
    
    
    cbcampaign_last = df4.groupby('campaign_last')['icb'].sum().to_frame().reset_index()
    cbcampaign_last = cbcampaign_last.rename(columns={'icb':'cb_campaign_last'})
    
    
    combinedcampaign_last = pd.merge(cbcampaign_last, totalcampaign_last, how='inner', left_on = 'campaign_last', right_on = 'campaign_last')
    combinedcampaign_last['cb_campaign_last_pc'] =  combinedcampaign_last['cb_campaign_last'] / combinedcampaign_last['total_campaign_last']
    
    combinedcampaign_last = combinedcampaign_last[['campaign_last', 'cb_campaign_last_pc']]
    
    df4 = df4.merge(combinedcampaign_last, how='inner', left_on = 'campaign_last', right_on = 'campaign_last')
    
    
    df4['cpu_name'] = df4['cpu_name'].fillna('nocpuname')
    
    totalcpu_name = df4.groupby('cpu_name')['icb'].count().to_frame().reset_index()
    totalcpu_name = totalcpu_name.rename(columns={'icb':'total_cpu_name'})
    
    
    cbcpu_name = df4.groupby('cpu_name')['icb'].sum().to_frame().reset_index()
    cbcpu_name = cbcpu_name.rename(columns={'icb':'cb_cpu_name'})
    
    
    combinedcpu_name = pd.merge(cbcpu_name, totalcpu_name, how='inner', left_on = 'cpu_name', right_on = 'cpu_name')
    combinedcpu_name['cb_cpu_name_pc'] =  combinedcpu_name['cb_cpu_name'] / combinedcpu_name['total_cpu_name']
    
    combinedcpu_name = combinedcpu_name[['cpu_name', 'cb_cpu_name_pc']]
    
    df4 = df4.merge(combinedcpu_name, how='inner', left_on = 'cpu_name', right_on = 'cpu_name')
    
    
    df4['gpu_name'] = df4['gpu_name'].fillna('nogpuname')
    
    totalgpu_name = df4.groupby('gpu_name')['icb'].count().to_frame().reset_index()
    totalgpu_name = totalgpu_name.rename(columns={'icb':'total_gpu_name'})
    
    
    cbgpu_name = df4.groupby('gpu_name')['icb'].sum().to_frame().reset_index()
    cbgpu_name = cbgpu_name.rename(columns={'icb':'cb_gpu_name'})
    
    
    combinedgpu_name = pd.merge(cbgpu_name, totalgpu_name, how='inner', left_on = 'gpu_name', right_on = 'gpu_name')
    combinedgpu_name['cb_gpu_name_pc'] =  combinedgpu_name['cb_gpu_name'] / combinedgpu_name['total_gpu_name']
    
    combinedgpu_name = combinedgpu_name[['gpu_name', 'cb_gpu_name_pc']]
    
    df4 = df4.merge(combinedgpu_name, how='inner', left_on = 'gpu_name', right_on = 'gpu_name')
    
    
    df4['web_vistype_browser'] = df4['web_vistype_browser'].fillna('nobrowser')
    
    totalbrowser = df4.groupby('web_vistype_browser')['icb'].count().to_frame().reset_index()
    totalbrowser = totalbrowser.rename(columns={'icb':'total_browser'})
    
    
    cbbrowser = df4.groupby('web_vistype_browser')['icb'].sum().to_frame().reset_index()
    cbbrowser = cbbrowser.rename(columns={'icb':'cb_browser'})
    
    
    combinedbrowser = pd.merge(cbbrowser, totalbrowser, how='inner', left_on = 'web_vistype_browser', right_on = 'web_vistype_browser')
    combinedbrowser['cb_browser_pc'] =  combinedbrowser['cb_browser'] / combinedbrowser['total_browser']
    
    combinedbrowser = combinedbrowser[['web_vistype_browser', 'cb_browser_pc']]
    
    df4 = df4.merge(combinedbrowser, how='inner', left_on = 'web_vistype_browser', right_on = 'web_vistype_browser')
    
    
    df4['os_name'] = df4['os_name'].fillna('noos')
    
    totalos = df4.groupby('os_name')['icb'].count().to_frame().reset_index()
    totalos = totalos.rename(columns={'icb':'total_os'})
    
    
    cbos = df4.groupby('os_name')['icb'].sum().to_frame().reset_index()
    cbos = cbos.rename(columns={'icb':'cb_os'})
    
    
    combinedos = pd.merge(cbos, totalos, how='inner', left_on = 'os_name', right_on = 'os_name')
    combinedos['cb_os_pc'] =  combinedos['cb_os'] / combinedos['total_os']
    
    combinedos = combinedos[['os_name', 'cb_os_pc']]
    
    df4 = df4.merge(combinedos, how='inner', left_on = 'os_name', right_on = 'os_name')
    
    df4['oid'] = df4['oid'].fillna('-100') 
    
    df4['oid'] = pd.to_numeric(df4['oid'], errors='coerce')
    df4['oid'] = df4['oid']/1000
    
    # tcvoc: Totl Campaign Visits of Customer
    df4['tcvoc'] = df4['campaign_visits']
    
    
    df4['ltv'] = df4['ltv'].fillna(0)
    
    df4['paid'] = df4['paid'].fillna(0)
    
    df4['order_id'] = df4['id_x']
    df4['order_id'] = pd.to_numeric(df4['order_id'], errors='raise')
    
    
    df4['survey_no'] = df4.groupby(['src_domain', 'item_name', 'provider'])['survey_no'].transform(lambda x: x.fillna(x.mean()))
    
    df4['survey_yes'] = df4.groupby(['src_domain', 'item_name', 'provider'])['survey_yes'].transform(lambda x: x.fillna(x.mean()))
    
    df4['paid_count'] = df4.groupby(['src_domain', 'item_name', 'provider'])['paid_count'].transform(lambda x: x.fillna(x.mean()))  
    
    df4['mail_sent'] = df4.groupby(['src_domain', 'item_name', 'provider'])['mail_sent'].transform(lambda x: x.fillna(x.mean())) 
    
    
    df4['same_campaign'] = df4.groupby(['src_domain', 'item_name', 'provider'])['same_campaign'].transform(lambda x: x.fillna(x.mean()))
    
    
    df5 = df4[['order_id', 'survey_no', 'survey_yes', 'paid_count', 'paid', 'mail_sent', 'ltv', 'oid', 'same_campaign', 'cpc', 'bgc', 'ikwn', 'tccbp', 'cb_tepp_pc', 'cb_item_pc', 'cb_provider_pc', 'cb_campaign_pc', 'cb_descriptor_pc', 'cb_referrer_pc', 'cb_src_domain_pc', 'cb_ai_pc', 'cb_ip_state_name_pc', 'cb_ip_city_pc', 'cb_item_key_pc', 'cb_strategy_id_pc', 'cb_at_pc', 'afalcts', 'cb_campaign_first_pc', 'cb_campaign_last_pc', 'cb_cpu_name_pc', 'cb_gpu_name_pc', 'cb_browser_pc', 'cb_os_pc', 'icb']]

    
    
    df5 = df5.set_index('order_id')
    
    
    
    
    df5['survey_no'] = df5.groupby('icb')['survey_no'].transform(lambda x: x.fillna(x.mean()))
    df5['survey_yes'] = df5.groupby('icb')['survey_yes'].transform(lambda x: x.fillna(x.mean()))
    df5['paid_count'] = df5.groupby('icb')['paid_count'].transform(lambda x: x.fillna(x.mean()))
    df5['mail_sent'] = df5.groupby('icb')['mail_sent'].transform(lambda x: x.fillna(x.mean()))
    df5['same_campaign'] = df5.groupby('icb')['same_campaign'].transform(lambda x: x.fillna(x.mean()))
    df5['cpc'] = df5.groupby('icb')['cpc'].transform(lambda x: x.fillna(x.mean()))

    
    
    
    return df5
    

    
    
    
    
    
    
    
    
    
# highcor: High Correlation Feature Detection     
    
def highcorr(df, treshold):
    
    df = df
    
    treshold = float(treshold) 
    
    corrcolls = set()
    corrmatrix = df.corr()
    
    for i in range(len(corrmatrix.columns)):
        for j in range(i):
            if abs(corrmatrix.iloc[i,j]) > treshold:
                colname = corrmatrix.columns[i]
                corrcolls.add(colname)
    
  
    
  
    
    
    print('There are highly correlated features detected in DataFrame')
    print()
    counter = 0
    for i in corrcolls:
        counter +=1
        print(counter, i)
    print()
    print()
    print(f'The {counter} features are highly correlated and creates coloniality. Please check VIF')
    print()
    print('---------------------------------------------------')
    print()
    #return corrcolls
   
    
    
    
    
    
# tma: The Modeler Algorithm
def tma(X,y, model):
    X = X
    y = y
    model = model 
    
   
    
    model = GaussianNB() 
    model.fit(X_train, np.ravel(y_train))
    
    y_pred = model.predict(X_test)
    predict = model.predict(X_test)
    
    y_true = np.array(y_test)
    y_pred = np.array(y_pred)
    
    
    
    
    
    the = y_test.reset_index()
    
    tdf = pd.DataFrame()
    tdf['order_id'] = the['order_id']
    tdf['real data'] = np.ravel(y_true)
    tdf['prediction'] = np.ravel(y_pred)
    
    
    
    GNBfeatureimportance = pd.DataFrame()
   
    GNBfeatureimportance['proba 1'] = np.ravel(model.predict_proba(X_test))[:round(X_test.shape[0]/2)]
    GNBfeatureimportance['proba 2'] = np.ravel(model.predict_proba(X_test))[round(X_test.shape[0]/2)-1:tdf.shape[0]]
    GNBfeatureimportance
    
    accuracy = accuracy_score(y_true, y_pred)
    
    recall = recall_score(y_true, y_pred)
    
    precision = precision_score(y_true, y_pred)
    
    f1 = f1_score(y_true, y_pred)
    
    tdf = tdf.set_index('order_id')
    
    print()
    print('The model has {:.2%} accuracy, {:.2%} recall, {:.2%} precision, {:.2%} f-1 score performance without deep KDD, complex modeling, advanced feature engineering and optimization.'.format(accuracy, recall, precision, f1))
    print()
    
    print('--------------------------------')
   
    print()
    
    return tdf
    
    
    
    
    
# cva: Cross Validation Algorithm
def cva(score, thenum, X, y):
    score = score
    thenum = thenum
    X = X
    y = y 
    
    GNB = GaussianNB()
    GNB.fit(X, np.ravel(y)) 
    
    y_pred = GNB.predict(X_test)
    predict = GNB.predict(X_test)
    
    
    y_true = np.array(y_test)
    y_pred = np.array(y_pred)
    
    
    GNBscores = cross_val_score(GNB, X, np.ravel(y), cv=thenum, scoring=score)
    
    GNBcv = pd.DataFrame()




    thelist = []

    for i in GNBscores:
        thelist.append(i)
    thelist

    roud = []

    for i in range(1, 11):
        roud.append(i)
    

    GNBcv['fold_bin'] = roud

    GNBcv['recall'] = thelist

    
    
    cvmean = np.mean(thelist)
    
    
    
    print()
    print('The {} fold Cross Validation shows that the model has {:.4} recall score which means with current structure the   model saves {:.2%} of Chargebacks; which is more than 50% (almost 60%) of the Chargebacks and Refunds'.format(thenum, cvmean, cvmean))
    print()
    
    print('--------------------------------')
    
 
    print()
    
    return GNBcv  
    
    
    
    
# cma: Confusion Matrix Algorithm

def cma(X,y,model):
    X = X
    y = y
    model = model 
    
   
    
    GBN = GaussianNB() 
    model = GaussianNB() 
    model.fit(X_train, np.ravel(y_train))
    
    y_pred = model.predict(X_test)
    predict = model.predict(X_test)
    
    y_true = np.array(y_test)
    y_pred = np.array(y_pred)
    
    
    
    accuracy = accuracy_score(y_true, y_pred)
    
    recall = recall_score(y_true, y_pred)
    
    precision = precision_score(y_true, y_pred)
    
    f1 = f1_score(y_true, y_pred)
    
    
    the = y_test.reset_index()
    
    tdf = pd.DataFrame()
    tdf['order_id'] = the['order_id']
    tdf['real data'] = np.ravel(y_true)
    tdf['prediction'] = np.ravel(y_pred)
    tdf = tdf.set_index('order_id')
    
    confusion_matrix(y_true, y_pred)
    
    predictions = model.predict(X_test)
    cm = confusion_matrix(y_test, predictions, labels=model.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap='Blues')

    
    TN = confusion_matrix(y_test, y_pred, labels=model.classes_)[0][0]
    FP = confusion_matrix(y_test, y_pred, labels=model.classes_)[0][1]
    FN = confusion_matrix(y_test, y_pred, labels=model.classes_)[1][0]
    TP = confusion_matrix(y_test, y_pred, labels=model.classes_)[1][1]
    
    
    
   
    
    
    
    print()
    print()
    print('     Gausian Naive Bayes Model Confusion Matrix')
    
    print()
    print()
    print('The model Confusion Matrix details:') 
    print()
    print('True Negative:   {}'.format(TN)) 
  
    print('False Positive:  {}'.format(FP))
  
    print('Falase Negative:  {}'.format(FN))
  
    print('True Positive:    {}'.format(TP))
    print()
    
    print('Note: The curret version of the model confusion rates are as given above and the recall is {:.2} '.format(recall))

    
    print()
    print('-----------------------------------------------------')
    print()
    
    
    return plt.show()
    
    
    
    
    
# tcr: The Classification report

def tcr(X, y):
    
    model = GaussianNB() 
    model.fit(X_train, np.ravel(y_train))
    
    y_pred = model.predict(X_test)
    predict = model.predict(X_test)
    
    y_true = np.array(y_test)
    y_pred = np.array(y_pred)
    
    
    
    accuracy = accuracy_score(y_true, y_pred)
    
    recall = recall_score(y_true, y_pred)
    
    precision = precision_score(y_true, y_pred)
    
    f1 = f1_score(y_true, y_pred)
    
    print()
    print('Model Classification Report:')
    print()
    print('The classification report shows that the current version of the model detects and predicts {:.0%} of the Refunds and  Chargebacks in real time, and predicts even before any single purchase.'.format(recall))
    print()
    print('------------------------------------------------------')
    print()
    print(classification_report(y_true, y_pred)) 
   
    
```
